# An example evaluation of results

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## ML models

Analysis of the accuracy

```python
df = pd.read_pickle("res/models_v2/quality.pkl.gz")


df = df.melt(id_vars=["model", "objective"], value_vars=[
             "test_score", "train_score"], value_name="score", var_name="dataset")
sns.catplot(
    data=df,
    kind="bar",
    x="model",
    y="score",
    hue="dataset",
    col="objective",
)
```

And the quality of solutions (regression models).

```python
df = pd.read_pickle("res/models_v2/learn.hw.pkl.gz")
display(df)

ml = ["RandomForrest", "SVC", "BayesianRidge"]

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ml, ax in zip(ml, axes):
    ax.set_title(ml)

    sns.scatterplot(
        data=df,
        x="hw_desired",
        y="hw_" + ml,
        hue="hw_train",
        ax=ax,
        s=10,
        alpha=0.4
    )

    ax.set(
        xlabel = "Vivado FPGA power",
        ylabel = "Estimated power",
        xlim = (0, 1.2),
        ylim = (0, 1.2)
    )

    ax.legend(title="Training data?")
```

# Final results

```python
df_hc = pd.read_json("res/hc_v2.eval.json.gz", orient="index")
df_hc["source"] = "hc_v2"
df_nsga = pd.read_json("res/nsga_v2.eval.json.gz", orient="index")
df_nsga["source"] = "nsga_v2"
df_random = pd.read_json("res/random.eval.json.gz", orient="index")
df_random["source"] = "random"

df = pd.concat([df_random, df_nsga, df_hc])
display(df)

sns.scatterplot(
    data=df,
    y="PSNR_AVG",
    x="power",
    hue="source",
    alpha=0.5

)
```
